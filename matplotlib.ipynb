```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=[1,2,3,4,5,6]
y=[34,44,55,67,88,99]
z=[33,44,55,66,77,88]
```

```python
plt.plot(x,y,'rD')
```

```python
plt.plot(x,y,color='green',marker='*',linestyle='',markersize=20)
```

```python
plt.plot(x,y,color='green',marker='+',linestyle='',alpha=1)
```

```python
plt.plot(x,y,label="A")
plt.plot(z,x,label="B")
plt.plot(y,z,label="C")
plt.xlabel("high")
plt.ylabel("low")
plt.title('multiple')
plt.legend(loc="best", shadow=True,fontsize="small")
```

```python
company=['hajk','hjakj','kkk','uuu']
revenue=[99,66,467,32]
Profit=[5,6,35,2]
```

```python
ypos=np.arange(len(company))
ypos
```

```python
xpos=np.arange(len(Profit))
xpos
```

```python
plt.bar(company,revenue)
```

```python
plt.xticks(ypos,company)
plt.ylabel("revenue(bin)")

plt.title("draw")
plt.bar(ypos,revenue,label="revenue")
plt.bar(ypos,Profit,label="Profit")
plt.legend()
```

```python
plt.xticks(ypos,company)
plt.ylabel("revenue(bin)")

plt.title("draw")
plt.bar(xpos-0.2,revenue,width=0.4,label="revenue")
plt.bar(xpos+0.2,Profit,width=0.4,label="Profit")
plt.legend()
```

```python
plt.xticks(ypos,company)
plt.ylabel("revenue(bin)")

plt.title("draw")
plt.barh(xpos-0.2,revenue,label="revenue")
plt.barh(xpos+0.2,Profit,label="Profit")
plt.legend()
```
